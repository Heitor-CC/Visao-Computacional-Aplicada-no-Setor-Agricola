```python
from ultralytics import YOLO
import torch
```

```python
#Verficar GPU
if torch.cuda.is_available():
    print(f"Usando GPU: {torch.cuda.get_device_name(0)}")
    device = 0
else:
    print("ATENÇÃO: GPU não detectada, usando CPU.")
    device = 'cpu'
```

```python

# yolov8n-seg.pt é o modelo pré-treinado para segmentação.
# utilizando inicalmente o modelo "nano", que é mais leve do que o "small" ou "medium"
model = YOLO('yolov8n-seg.pt') 

# treinamento
print("\nIniciando treinamento com o dataset ROCOLE")

model.train(
    data='../datasets/rocole_seg/data.yaml',  # Aponta para o dataset 
    epochs=50,             # quantas vezes roda
    imgsz=640,             # tamanho da imagem (padrão da YOLO)
    device=device,         # gpu ou cpu
    batch=8,              # quantas fotos processa por vez 16 - 8 - 4
    patience=10,           # se não melhorar por 10 épocas, para o funcionamento
    name='rocole_modelo_seg' # Nome da pasta onde vai salvar o resultado
)

print("treinamento finalizado!")
```
